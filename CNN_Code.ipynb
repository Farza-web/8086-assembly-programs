```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import tensorflow as tf#for Deep Learning we use tenser flow
from  tensorflow .keras.models import Sequential#API of Deep Learning
from tensorflow.keras.layers import Dense,Activation,Flatten,Conv2D,MaxPooling2D
import pickle#image apply kore cnn apply korbo nijer image diye MaxPooling2D apply korbo
```

```python
img=np.random.rand(128,128,3)
```

```python
plt.imshow(img)
plt.title("Random Image")
plt.axis("Off")
plt.show()
```

```python
model = Sequential([
    Conv2D(filters=8, kernel_size=(3,3), activation="relu", input_shape=(128,128,3)),
    MaxPooling2D(pool_size=(2,2)),

    Conv2D(filters=8, kernel_size=(3,3), activation="relu"),
    MaxPooling2D(pool_size=(2,2)),

    Flatten(),

    Dense(units=1, activation="sigmoid")
])
```
